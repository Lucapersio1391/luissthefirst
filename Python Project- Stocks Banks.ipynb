```python
start = datetime.datetime(2016, 1, 1)# the starting point of the data, taking into account the year, the month and the day
end = datetime.datetime(2018, 1, 1)# the ending point of the data,
```

** We import the data of the banks taking into account the starting point and the ending point wrote above. Infact we pass to datareader the name of the bank  and the source iex as a string to grab the data, and after the start and the end for time information**


```python

# Bank of America
BAC = data.DataReader("BAC", 'iex', start, end)

# CitiGroup
C = data.DataReader("C", 'iex', start, end)

# Goldman Sachs
GS = data.DataReader("GS", 'iex', start, end)

# JPMorgan Chase
JPM = data.DataReader("JPM", 'iex', start, end)

# Morgan Stanley
MS = data.DataReader("MS", 'iex', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'iex', start, end)
```

** We create a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

** by using pd.concat we concatenate the bank dataframes together to a single data frame called bank_stocks setting the keys argument equal to the tickers list. **

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC],axis=1,keys=tickers)# here we concatenate all the dataframes together along the columns
```

** Set the column name levels**

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

** Check the head of the bank_stocks dataframe.**

```python
bank_stocks.head()
```

** Here we compute the max close price for each bank's stocks throughout the time period **

```python
bank_stocks.xs(key='close',axis=1,level='Stock Info').max()# we do the cross section on 'Close', and i want to do it on columns so axis=1. i specifeid Stock info as level.
# So in summary i want the max of the Close columns of all the banks.
```

 ** Now we create an empy data frame called returns.This dataframe will contains the reurns for each banks's stock. 
Rreturns are typically defined by:

 rt = (pt-pt-1)/pt-1

So it is the price at time t less the price the time at t-1 divided by the preice at time t-1.**

```python
returns = pd.DataFrame()
```

** We can use pandas pct_change() method on the 'close' column to create a column representing this return value. So we do a for loop that for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['close'].pct_change()
returns.head()
```

** The first values is NaN because you can't see the return in the first day 
because there isn't anything in the past as reference. **

**  We create a pairplot using seaborn of the returns dataframe. **

```python
import seaborn as sns
sns.set_style('whitegrid')
sns.pairplot(returns[1:])
# we need to spceify returns[1:] because we wont everything starting from position 1
# because the first row is NaN.
```

** Now using the returns DataFrame the objective is to figure out on what dates each bank stock had the best and worst single day returns. Starting with the worst day we see that for all the banks there is the same day for the worst drop.
We compute it by using idxmin that shows us all the data for the banks with the worst value for the stocks.**

```python
# Worst Drop 
returns.idxmin()
```

** In the same way we compute the max value and we see the best day for the stoks of the different  banks and we can see that the Bank of America and the city Group Bank registered the maximum return for the stoks in the same day.**

```python
# Max Return
returns.idxmax()
```

 ** Here we compute the the stadard deviation for the stocks in order to understand which price of the stoks have a big variation. Of course more the standard deviation is high  more the stock is risky.**

```python
returns.std() # Citigroup riskiest
```

```python
returns.loc['2017-01-01':'2017-12-31'].std() # Very similar risk profiles, The highest are for BAC and MS.
```

** Create a distplot using seaborn of the 2017 returns for Morgan Stanley **

```python
sns.distplot(returns.loc['2017-01-01':'2017-12-31']['MS Return'],color='green',bins=50)
```

** Create a distplot using seaborn of the 2017 returns for BAC Return **

```python
sns.distplot(returns.loc['2017-01-01':'2017-12-31']['BAC Return'],color='red',bins=50)
```

** We create a line plot showing 'close' price for each bank for the entire index of time. We use two different methods, the first one is the 'For Loop' Method and the second one is the 'Cross Section Method'.**

# Correlation Map


** The first thing to do is to compute the correlation of the closing price of the different Banks **

```python
bank_stocks.xs(key='close',axis=1,level='Stock Info').corr()
```

** There are high and positive correlation among all the bank stocks **

** We Create a heatmap of the correlation between the stocks Close Price.**

```python
sns.heatmap(bank_stocks.xs(key='close',axis=1,level='Stock Info').corr(),annot=True)
```

** We use seaborn's clustermap to cluster the correlations together:**

```python
sns.clustermap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```

** From the clustermap we can se that there are two different groups of banks. the first one is composed by Bank of America, City Group and Morgan Stanly. the second one is composed by JP Morgan, Goldman Sakcs and Wells Fargo. In the second group we can observe a strong correlation between Golman Saks and Jp Morgan.There is also a strong Correlation beteen Morgan Stanlley and Bank of America **
